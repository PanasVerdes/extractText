```python
import pdfplumber  # Importa la biblioteca pdfplumber

def encontrar_pagina_justificacion(pdf_path):
    max_length = 0  # Inicializa la longitud máxima a 0
    pagina_justificacion = None  # Inicializa la página de justificación como nula
    
    with pdfplumber.open(pdf_path) as pdf:  # Abre el archivo PDF
        for i, page in enumerate(pdf.pages, 1):  # Itera sobre todas las páginas del PDF
            text = page.extract_text()  # Extrae el texto de la página
            parrafos = text.split('\n\n')  # Divide el texto en párrafos
            for parrafo in parrafos:  # Itera sobre cada párrafo
                longitud_justificada = sum(1 for c in parrafo if c.isalnum() or c.isspace())  # Calcula la longitud justificada del párrafo
                if longitud_justificada > max_length:  # Si la longitud justificada es mayor que la longitud máxima encontrada hasta ahora
                    max_length = longitud_justificada  # Actualiza la longitud máxima
                    pagina_justificacion = i  # Actualiza la página de justificación
    
    return pagina_justificacion  # Devuelve el número de página de la justificación


pdf_path = "/kaggle/input/ketodiet/Doc_Alumnos/Doc_Alumnos/Proyecto de investigación_TESIS_Jonathan Orduña_opción a liberar.pdf"  # Ruta del archivo PDF
pagina_justificacion = encontrar_pagina_justificacion(pdf_path)  # Encuentra la página de la justificación en el PDF

if pagina_justificacion:  # Si se encontró la página de justificación
    print("La justificación se encuentra en la página:", pagina_justificacion)  # Imprime el número de página
else:  # Si no se encontró la página de justificación
    print("No se encontró justificación en el PDF.")  # Imprime un mensaje indicando que no se encontró la justificación
```

```python
import pdfplumber  # Importa la biblioteca pdfplumber

def encontrar_palabra_en_documento(pdf_path, palabra):
    paginas_con_palabra = []  # Lista para almacenar las páginas que contienen la palabra
    with pdfplumber.open(pdf_path) as pdf:  # Abre el archivo PDF
        for i, page in enumerate(pdf.pages, 1):  # Itera sobre todas las páginas del PDF
            text = page.extract_text()  # Extrae el texto de la página
            if palabra in text:  # Verifica si la palabra está en el texto de la página
                paginas_con_palabra.append(i)  # Agrega el número de página a la lista
    return paginas_con_palabra  # Devuelve la lista de páginas que contienen la palabra

# Ruta del archivo PDF y palabra a buscar
pdf_path = "/kaggle/input/ketodiet/Doc_Alumnos/Doc_Alumnos/Proyecto de investigación_TESIS_Jonathan Orduña_opción a liberar.pdf"
palabra_a_buscar = "Justificación"

# Encuentra las páginas que contienen la palabra en el PDF
paginas_palabra = encontrar_palabra_en_documento(pdf_path, palabra_a_buscar)

if paginas_palabra:  # Si se encontraron páginas con la palabra
    print(f"La palabra '{palabra_a_buscar}' se encuentra en las siguientes páginas:", paginas_palabra)  # Imprime las páginas encontradas
else:  # Si no se encontró la palabra en el PDF
    print(f"No se encontró la palabra '{palabra_a_buscar}' en el PDF.")  # Imprime un mensaje indicando que no se encontró la palabra
```

```python
import csv
import pdfplumber

def encontrar_justificacion_en_documento(pdf_path, csv_output):
    justificaciones = []  # Lista para almacenar las justificaciones encontradas
    with pdfplumber.open(pdf_path) as pdf:  # Abre el archivo PDF
        for i, page in enumerate(pdf.pages, 1):  # Itera sobre todas las páginas del PDF
            if i > 5:  # Ignora las primeras 5 páginas
                text = page.extract_text()  # Extrae el texto de la página
                parrafos = text.split('\n\n')  # Divide el texto en párrafos
                for parrafo in parrafos:  # Itera sobre cada párrafo
                    if "Justificación" in parrafo:  # Verifica si el párrafo contiene "Justificación"
                        justificaciones.append(parrafo.strip())  # Agrega la justificación a la lista

    # Guardar las justificaciones en un archivo CSV
    with open(csv_output, 'w', newline='', encoding='utf-8') as csvfile:  # Abre el archivo CSV para escritura
        writer = csv.writer(csvfile)  # Inicializa el escritor CSV
        writer.writerow(['Página', 'Justificación'])  # Escribe el encabezado del CSV
        for i, justificacion in enumerate(justificaciones, 6):  # Enumera a partir de la página 6
            writer.writerow([i, justificacion])  # Escribe cada justificación junto con el número de página

# Ruta del archivo PDF y archivo CSV de salida
pdf_path = "/kaggle/input/ketodiet/Doc_Alumnos/Doc_Alumnos/Tesis_Luis_Armando_Gutierrez_Jaime_U1903058M0065.pdf"
csv_output = "justificaciones.csv"

# Llama a la función para encontrar las justificaciones y guardarlas en el archivo CSV
encontrar_justificacion_en_documento(pdf_path, csv_output)
print(f"Se han guardado las justificaciones en el archivo CSV: {csv_output}")  # Imprime un mensaje indicando la finalización del proceso
```
